```python
# TODOリストの本体
todo_list = []

def show_list():
    """TODO一覧を番号付きで表示（1始まり）"""
    print("【TODOリスト】")
    if not todo_list:
        print("(空です)")
        return
    for i, item in enumerate(todo_list, start=1):
        print(f"{i}. {item}")

def add_item():
    task = input("追加する項目: ").strip()
    if task:
        todo_list.append(task)

def delete_item():
    """仕様の削除機能（3）"""
    print("3:削除")
    if not todo_list:
        print("【TODOリスト】")
        print("(空です)")
        return

    # 現在のリストを表示
    show_list()

    # 番号入力を受け付け
    s = input("削除する番号: ").strip()

    # 数字以外はエラー
    if not s.isdigit():
        print("無効な番号です。")
        return

    n = int(s)

    # 範囲外はエラー
    if n < 1 or n > len(todo_list):
        print("無効な番号です。")
        return

    # 正常時：対象を削除
    todo_list.pop(n - 1)

def save_and_exit():
    # 必要に応じて保存先を変更してください
    with open("todo.txt", "w", encoding="utf-8") as f:
        for item in todo_list:
            f.write(item + "\n")
    print("TODOリストを保存して終了します.")

# メインループ
while True:
    choice = input("番号を入力してください[1:追加、2:表示、3:削除、4:保存と終了]:")
    if choice == "1":
        print("1:追加")
        add_item()
    elif choice == "2":
        print("2:表示")
        show_list()
    elif choice == "3":
        delete_item()
    elif choice == "4":
        print("4:保存と終了")
        save_and_exit()
        break
    else:
        print("無効な番号です。")
```
